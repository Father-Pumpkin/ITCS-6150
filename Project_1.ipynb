```python
import random
```

```python
class PuzzleBoard:
    board = [[],[],[]]
    rowLocationOfEmpty = -1
    colLocationOfEmpty = -1
    emptySymbol = -1;
    def __init__(self, *args):
        if(len(args) == 0):
            self.board = [[1,2,3],[4,5,6],[7,8,0]]
        else:
            for i in range(len(args)):
                for j in range(len(args[0])):
                    if args[i][j] == 0 or args[i][j] == ' ' or args[i][j] == '':
                        self.rowLocationOfEmpty = i
                        self.colLocationOfEmpty = j
                        self.emptySymbol = args[i][j]
            self.board = [args[0], args[1], args[2]]
    
    # directionToMove 1,2,3,4 => Up, Down, Left, Right
    def move(self, directionToMove):
        # 1,2,3,4 => Up, Down, Left, Right || -1 is an error, piece impossible to move
        directionToMove = self.checkValidMove(directionToMove)
        if directionToMove == -1:
            return -1;
        #set the row/col of the piece that's getting swapped
        if directionToMove == 1:
            row = self.rowLocationOfEmpty - 1
            col = self.colLocationOfEmpty
        if directionToMove == 2:
            row = self.rowLocationOfEmpty + 1
            col = self.colLocationOfEmpty
        if directionToMove == 3:
            row = self.rowLocationOfEmpty
            col = self.colLocationOfEmpty - 1
        if directionToMove == 4:
            row = self.rowLocationOfEmpty
            col = self.colLocationOfEmpty + 1
        self.board[self.rowLocationOfEmpty][self.colLocationOfEmpty] = self.board[row][col]
        self.board[row][col] = self.emptySymbol
        self.rowLocationOfEmpty = row
        self.colLocationOfEmpty = col
    
    def prettyPrint(self):
        for i in range(len(self.board)):
            for j in range(len(self.board[0])):
                print(self.board[i][j], end =" ")
            print('')
        print()
        
    def printLocationOfEmpty(self):
        print("Empty is at:", self.rowLocationOfEmpty, self.colLocationOfEmpty)
        
    # 1,2,3,4 => Up, Down, Left, Right
    # @return directionToMove for valid || -1 is an error, piece impossible to move
    def checkValidMove(self, directionToMove):
        if directionToMove == 1:
            if self.rowLocationOfEmpty == 0:
                return -1
            else:
                return 1
        if directionToMove == 2:
            if self.rowLocationOfEmpty == 2:
                return -1
            else:
                return 2
        if directionToMove == 3:
            if self.colLocationOfEmpty == 0:
                return -1
            else:
                return 3
        if directionToMove == 4:
            if self.colLocationOfEmpty == 2:
                return -1
            else:
                return 4
    def checkCorrectness(self, goal):
        correctnessValue = 0;
        for i in range(len(self.board)):
            for j in range(len(self.board[0])):
                if self.board[i][j] == goal.board[i][j]:
                    correctnessValue += 1
        return correctnessValue
```

```python
class Search:
    currentBoard = [[],[],[]]
    goalBoard = [[],[],[]]
    checkedBoards = []
    def __init__(self, *args):
        self.currentBoard = args[0]
        self.goalBoard = args[1]
    
    def takeBestMove(self):
        bestMove = random.randint(1,4);
        mostCorrect = self.currentBoard.checkCorrectness(self.goalBoard);
        if self.currentBoard.checkValidMove(1) != -1:
            self.currentBoard.move(1)
            if self.currentBoard.board not in self.checkedBoards and self.currentBoard.checkCorrectness(self.goalBoard) > mostCorrect:
                bestMove = 1
                mostCorrect = self.currentBoard.checkCorrectness(self.goalBoard)
            self.currentBoard.move(2)
        if self.currentBoard.checkValidMove(2) != -1:
            self.currentBoard.move(2)
            if self.currentBoard.board not in self.checkedBoards and self.currentBoard.checkCorrectness(self.goalBoard) > mostCorrect:
                bestMove = 2
                mostCorrect = self.currentBoard.checkCorrectness(self.goalBoard)
            self.currentBoard.move(1)
        if self.currentBoard.checkValidMove(3) != -1:
            self.currentBoard.move(3)
            if self.currentBoard.board not in self.checkedBoards and self.currentBoard.checkCorrectness(self.goalBoard) > mostCorrect:
                bestMove = 3
                mostCorrect = self.currentBoard.checkCorrectness(self.goalBoard)
            self.currentBoard.move(4)
        if self.currentBoard.checkValidMove(4) != -1:
            self.currentBoard.move(4)
            if self.currentBoard.board not in self.checkedBoards and self.currentBoard.checkCorrectness(self.goalBoard) > mostCorrect:
                bestMove = 4
                mostCorrect = self.currentBoard.checkCorrectness(self.goalBoard)
            self.currentBoard.move(3)
        print("Taking Move", bestMove)
        self.checkedBoards.append(self.currentBoard.board)
        self.currentBoard.move(bestMove)
```

```python
pb = PuzzleBoard([0,1,3], [4,2,5],[7,8,6])
pbGoal = PuzzleBoard([1,2,3], [4,5,6],[7,8,0])
pb.prettyPrint()
pbGoal.prettyPrint()
print(pb.checkCorrectness(pbGoal))
s = Search(pb, pbGoal)
s.takeBestMove()
pb.prettyPrint()
print(pb.checkCorrectness(pbGoal))
s.takeBestMove()
pb.prettyPrint()
print(pb.checkCorrectness(pbGoal))
s.takeBestMove()
pb.prettyPrint()
print(pb.checkCorrectness(pbGoal))
s.takeBestMove()
pb.prettyPrint()
print(pb.checkCorrectness(pbGoal))
s.takeBestMove()
pb.prettyPrint()
print(pb.checkCorrectness(pbGoal))
s.takeBestMove()
pb.prettyPrint()
print(pb.checkCorrectness(pbGoal))

```
