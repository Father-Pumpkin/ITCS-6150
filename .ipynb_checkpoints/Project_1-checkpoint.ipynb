```python
class PuzzleBoard:
    board = [[],[],[]]
    rowLocationOfEmpty = -1
    colLocationOfEmpty = -1
    emptySymbol = -1;
    def __init__(self, *args):
        if(len(args) == 0):
            self.board = [[1,2,3],[4,5,6],[7,8,0]]
        else:
            for i in range(len(args)):
                for j in range(len(args[0])):
                    if args[i][j] == 0 or args[i][j] == ' ' or args[i][j] == '':
                        self.rowLocationOfEmpty = i
                        self.colLocationOfEmpty = j
                        self.emptySymbol = args[i][j]
            self.board = [args[0], args[1], args[2]]

    # pieceToMove is an int 0-8 of the location of the pieceToMove
    def move(self, pieceToMove):
        row = pieceToMove//3
        col = pieceToMove%3
        # 1,2,3,4 => Up, Down, Left, Right || -1 is an error, piece impossible to move
        directionToMove = self.__findValidMove(pieceToMove)
        if directionToMove == -1:
            print("Invalid move")
            return -1;
        self.board[self.rowLocationOfEmpty][self.colLocationOfEmpty] = self.board[row][col]
        self.board[row][col] = self.emptySymbol
        self.rowLocationOfEmpty = row
        self.colLocationOfEmpty = col
    
    def prettyPrint(self):
        for i in range(len(self.board)):
            for j in range(len(self.board[0])):
                print(self.board[i][j], end =" ")
            print('')
        print()
        
    def printLocationOfEmpty(self):
        print("Empty is at:", self.rowLocationOfEmpty, self.colLocationOfEmpty)
        
    # 1,2,3,4 => Up, Down, Left, Right || -1 is an error, piece impossible to move
    def __findValidMove(self, pieceToMove):
        #get location of pieceToMove
        row = pieceToMove//3
        col = pieceToMove%3
        # can't move the empty piece
        if self.board[row][col] == self.emptySymbol:
            return -1
        # if empty piece is in the top row
        if self.rowLocationOfEmpty == 0:
            if col == self.colLocationOfEmpty and row == 1:
                return 1
            
        # if empty piece is in the bottom row
        if self.rowLocationOfEmpty == 2:
            if col == self.colLocationOfEmpty and row == 1:
                return 2
        
        # if empty piece is in the leftmost column
        if self.colLocationOfEmpty == 0:
            if row == self.rowLocationOfEmpty and col == 1:
                return 3
        
        # if empty piece is in the rightmost column
        if self.colLocationOfEmpty == 2:
            if row == self.rowLocationOfEmpty and col == 1:
                return 4
            
        # if empty piece is in the middle row
        if self.rowLocationOfEmpty == 1:
            if self.colLocationOfEmpty == col:
                if row == 0:
                    return 2
                if row == 2:
                    return 1
            if row == 1:
                if self.colLocationOfEmpty == col-1:
                    return 3
                if self.colLocationOfEmpty == col+1:
                    return 4
        
        # otherwise no valid move for that piece
        return -1
        
```

```python
pb = PuzzleBoard([0,1,3], [4,2,5],[7,8,6])
pb.prettyPrint()
pb.move(1)
pb.prettyPrint()
pb.move(4)
pb.prettyPrint()
pb.move(5)
pb.prettyPrint()
pb.move(8)
pb.prettyPrint();

```
